```python
import importlib
import Colony_Analysis as ca
#import Plate_Shuffling as ps
from matplotlib import pyplot as plt
import BnA_Correction as ba #Boone and Andrew's lab correction
import os
import numpy as np
import pandas as pd

from skimage.color import rgb2gray
import skimage.io
from skimage.transform import rotate
```

```python
#create the plate,row,col coordinate system to store the data, we used 1536 (32rows, 48 columns) plates
plate_map = ca.Generate_Plate_Map('/Users/paucac/Documents/PhD/CAB/Yeast Libraries/Tef2_mCherry_SWAT/OEx_plate_map.txt')

#create the plate,row,col coordinate system
n_row=32
n_col=48
n_plate=16
grid_size='32-48'
row = np.arange(0, n_row)
col = np.arange(0, n_col)
col,row = np.meshgrid(col,row)
plate= np.arange(0,n_plate)
row_col = np.vstack((row.flatten(),col.flatten())).T
plate=np.repeat(plate,n_row * n_col).reshape(-1,1)
row_col =np.tile(row_col.T,n_plate).T
coordinate= np.hstack([plate, row_col])

#create the gene name column !! need to use the right plate map
gene_col=[] #stores the gene ID
for p in range(n_plate):
    gene_col.append(plate_map[p+1].flatten())
gene_col = np.hstack(gene_col).reshape(-1,1)

#create the data frame storing a reading for each individual position
Data=pd.DataFrame(data=np.hstack((gene_col,coordinate)), columns=['gene','plate', 'row', 'column'])
Data['plate']=Data['plate'].astype('int64')
Data['row']=Data['row'].astype('int64')
Data['column']=Data['column'].astype('int64')


```

```python
#extract size DATA from SC plates in white light 
plot=0

# set some Pyphe paramters
#s,t,d,i >>size_threshold parameter/ image back_ground threshold parameter/ d?/invert image
#background thresh the lower the more the background is sensed as colony
params= [0.05,1,2, 1] #s,t,d,i 


time='48' 
plate_path_dict={}
image_path = 'raw_data/plate_images/screen20_23.3.15/white/'

for filename in os.listdir(image_path):
    if '.Tif' in filename or '.png' in filename:
        i = filename.index('p') + 1
        plate_num = int(filename[i:i+2])  
        plate_path_dict[plate_num]  = image_path+ '/' + filename
#initialize the object storing our data
Area= np.zeros((32,48,n_plate))
s,t,d, negate = params
for p, plate in plate_path_dict.items():           
    print(plate)

    im=skimage.io.imread(plate_path_dict[p])[460:1432,530:1990]
    im = rgb2gray(im)


    #extract 
    plate_data, QC = ca.quantify_single_image_size(im, grid_size, True, t, d, s, negate)
    for i in range(len(plate_data)):
        r = int(plate_data.iloc[i,5])
        c = int(plate_data.iloc[i,6])
        Area[r-1,c-1,p-1] = plate_data.iloc[i,1]

    if plot==1:
        plt.figure(figsize=(20,15))

        plt.subplot(1,3,1)
        plt.imshow(QC[:,:,:]*1)
        plt.subplot(1,3,2)
        plt.imshow(Area[:,:,p-1] )
        plt.show()


Data['area.'+time ] = Area.transpose(2,0,1).flatten()
# Data_SCGN['mean_intensity.'+time] = Mean_intensity.transpose(2,0,1).flatten()
```

```python
%%time
#extract fluo DATA from SC plates in fluo light. average fluo obtained from pictures obtained 
#from 0,90,180,270 degree angles

plot=0

#adjust parameters for different time points:

params = [0.5,0.7,3, 0]


rotations={'A':0, 'B':270, 'C':180,'D':90}
time ='48'
plate_path_dict={}
image_path = 'raw_data/plate_images/screen20_23.3.15/fluo/'
for filename in os.listdir(image_path):
    if '.Tif' in filename or '.png' in filename:
        i = filename.index('p') + 1

        plate_num = int(filename[i:i+2])  
        rotation = filename[i+2:i+3]
        #store plate name as a nested dict: plate_path_dict[plate number] :{A:, B:,C:,D:}
        if plate_num in  plate_path_dict.keys():
            plate_path_dict[plate_num][rotation]=image_path+ '/' + filename
        else:
            plate_path_dict[plate_num]={rotation: image_path+ '/' + filename}

s,t,d, negate = params
for rotation in ['A','B','C','D']:
     # initialize the object storing our data
    Mean_intensity = np.zeros((32,48,n_plate ))
    for p, plates in plate_path_dict.items(): 
            print(p, rotation, plates[rotation])
            im_fluo=skimage.io.imread(plates[rotation])
            #rotate the plate :
            im_fluo= rotate(im_fluo,rotations[rotation])

            im_fluo,[angle,top,bottom,left,right]=ca.Velber_Preprocess( im_fluo, crop=[0,0,0,0])

            #now that im_fluo is cropped (lighter) we polish the alignment
                #now that the image is lighter we polish the alignment:
            im_fluo,a = ca.Rotate_Plate(im_fluo , 0.2)
            angle += a
            im_fluo ,a = ca.Rotate_Plate(im_fluo , 0.1)
            angle += a
            im_fluo ,a = ca.Rotate_Plate(im_fluo , 0.05)
            angle += a
            ## the whole colony extraction pipeline expects the background to be constant. Should make sure that we 
            #trim so that none agar surfaces are removed
            #im_fluo = im_fluo[15:-15,15:-15]

            #extract 
            plate_data, QC = ca.quantify_single_image_size(im_fluo, grid_size, True, t, d, s, negate)
            for i in range(len(plate_data)):
                r = int(plate_data.iloc[i,5])
                c = int(plate_data.iloc[i,6])
                Mean_intensity[r-1,c-1,p-1]  = plate_data.iloc[i,3]
            if plot==1:
                plt.figure(figsize=(10,10))

                plt.subplot(1,2,1)
                plt.imshow(QC[:,:,:]*2)#
                plt.subplot(1,2,2)
                plt.imshow(Mean_intensity[:,:,p-1] )#
                plt.show()


    Data['mean_intensity.'+time +'.'+str(rotation)] = Mean_intensity.transpose(2,0,1).flatten()
#we average over all plates angles
Data['mean_intensity.'+str(time)]=Data.loc[:,['mean_intensity.'+str(time)+'.A','mean_intensity.'+str(time)+'.B','mean_intensity.'+str(time)+'.C','mean_intensity.'+str(time)+'.D']].mean(axis=1)
   
```

```python
%%time

#turn 0 into NaNs
metrics = [m for m in Data.columns if 'area' in m or 'intensity'  in m   ]
Data[metrics] = Data[metrics].replace(0, np.nan)

##perfrom the Boone and Andrews lab correction:

########## (N1) Plate normalization ##########
metrics=['mean_intensity.48','area.48']
Data_BA_corrected = ba.N1_normalisation(Data, metrics ,default_cols=['gene','plate','row','column'], p_format=[32,48] )

########## (N2) Spatial normalization ##########
metrics=[metric for metric in Data_BA_corrected.columns if 'corrected' in metric]
default_cols=['gene','plate','row','column','mean_intensity.48','area.48']
Data_BA_corrected = ba.N2_spatial_normalisation(Data_BA_corrected, metrics ,default_cols=default_cols, p_format=[32,48])

########## (N3) Row column effect normalization ##########
metrics=[metric for metric in Data_BA_corrected.columns if 'corrected' in metric]
default_cols=['gene','plate','row','column','mean_intensity.48','area.48']

Data_BA_corrected = ba.N3_rowcol_Lowess(Data_BA_corrected, metrics, default_cols)

######### (F3) Jackknife filter  ##########
metrics=[metric for metric in Data_BA_corrected.columns if 'corrected' in metric]
default_cols=['gene','plate','row','column','mean_intensity.48','area.48']

Data_BA_corrected = ba.F3_jacknife_filter(Data_BA_corrected, metrics, default_cols)

########## (N4) Plate normalization 2 ##########
metrics=[metric for metric in Data_BA_corrected.columns if 'corrected' in metric]
default_cols=['gene','plate','row','mean_intensity.48','area.48']

Data_BA_corrected = ba.N2_spatial_normalisation(Data_BA_corrected, metrics ,default_cols=default_cols, p_format=[32,48])

#Filter genes with less than 3 colonies
metrics=[c for c in Data_BA_corrected.columns if 'area' in c or 'mean_intensity' in c ]

print(Data_BA_corrected[metrics].isna().sum())
Data_BA_corrected = ca.Filter_3(Data_BA_corrected,  metrics)

print(Data_BA_corrected[metrics].isna().sum())
```

```python

# here we group by_gene the filtered replicates, this computes the mean, std, and count 
#the replicate number for each metric
metrics=[c for c in Data_BA_corrected.columns if 'area' in c or 'mean_intensity' in c or 'plate' in c ]
GA20 = ca.group_by_gene(Data_BA_corrected, metrics)# save to csv
```

```python
#produce middle figure



DF = GA20.copy()

metrics = ['corrected_mean_intensity.48','corrected_area.48']

for metric in metrics:
        #we first correct the std using the screens mean
        #we then correct the mean
        DF[metric +'_std'] = (DF[metric+'_std'])/DF[metric+'_mean'].std()
        DF[metric +'_mean'] = (DF[metric+'_mean']-DF[metric+'_mean'].mean())/DF[metric+'_mean'].std()
        
        

from scipy.stats import mannwhitneyu
metric = 'corrected_mean_intensity.48'



i_nan = DF[metric+'_mean'].isna() ==0

i_sort =  DF[i_nan][metric+'_mean'].argsort()[::-1]
i_sort_top =  (DF[i_nan][metric+'_mean']-0*DF[i_nan][metric+'_std']).argsort()[::-1]
i_sort_bottom =  (DF[i_nan][metric+'_mean']+0*DF[i_nan][metric+'_std']).argsort()[::-1]

n_genes= 6#int(0.02*(DF[ metric+'_mean'].isna()==0).sum())
i_middle_group =  np.arange(len( DF[i_nan])/2-n_genes/2,len( DF[i_nan])/2+n_genes/2+2)
i_top_group =  np.arange(0,n_genes)
i_bottom_group =  np.arange(-n_genes,0)

i_all= np.arange(i_nan.sum())
All= DF[i_nan].iloc[i_sort][metric+'_mean']
All_err= DF[i_nan].iloc[i_sort][metric+'_std']

Top = DF[i_nan].iloc[i_sort_top].iloc[i_top_group][metric+'_mean']
Top_err = DF[i_nan].iloc[i_sort_top].iloc[i_top_group][metric+'_std']
Top_genes = DF[i_nan].iloc[i_sort_top].iloc[i_top_group]['gene'].values

#Top_genes = DF[i_nan].iloc[i_sort_top].iloc[i_top_group]['gene'].values
Middle =  DF[i_nan].iloc[i_sort].iloc[i_middle_group][metric+'_mean']
Middle_err = DF[i_nan].iloc[i_sort].iloc[i_middle_group][metric+'_std']
Middle_genes =  DF[i_nan].iloc[i_sort].iloc[i_middle_group]['gene'].values
                                                                    
Bottom = DF[i_nan].iloc[i_sort_bottom].iloc[i_bottom_group][metric+'_mean']
Bottom_err = DF[i_nan].iloc[i_sort_bottom].iloc[i_bottom_group][metric+'_std']
Bottom_genes = DF[i_nan].iloc[i_sort_bottom].iloc[i_bottom_group]['gene'].values

i_WT= DF[i_nan].iloc[i_sort].gene =='WT'
WT_score=DF[i_nan].iloc[i_sort][i_WT][metric+'_mean']
WT_err=DF[i_nan].iloc[i_sort][i_WT][metric+'_std']

Bottom_color=[106/255,213/255,255/255]
Middle_color=[64/255,150/255,234/255]
Top_color=[0/255,77/255,128/255]




H=5#figure hight
fig, axs = plt.subplots(figsize=(H*1.3,H))
i = list(set(np.arange(0,len(i_all), 1))-set(i_top_group)-set(i_bottom_group)-set(i_middle_group))
axs.errorbar(i_all[i], All.iloc[i], All_err.iloc[i], linestyle='', marker = '.', color='k', alpha=0.1)
#axs.errorbar(i_all[i], All.iloc[i], All_err.iloc[i], linestyle='', marker = '_', color='k', alpha=0.08)
axs.errorbar(i_middle_group, Middle, Middle_err, c=Middle_color , marker = 'o',  markersize=H*1.5,  capsize=H, elinewidth=H, markeredgewidth=H/2, alpha=1)
axs.errorbar(i_top_group,Top, Top_err, c=Top_color, marker = 'o',  markersize=H*1.5,  capsize=H, elinewidth=H, markeredgewidth=H/2, alpha=1 )
axs.errorbar(i_bottom_group+len(i_all),Bottom, Bottom_err, c=Bottom_color, marker = 'o',  markersize=H*1.5, capsize=H, elinewidth=H, markeredgewidth=H/2 , alpha=1)
axs.errorbar(np.where(i_WT)[0], WT_score,WT_err, linestyle='', marker = 'o',  markersize=H*1.5, color='red' ,  capsize=H, elinewidth=H, markeredgewidth=H/2, alpha=1)


axs.set_title('Top, Middle and Bottom Ranks',fontsize=H*2.5)
axs.set_xlabel('Gene Rank',fontsize=H*2.5)
axs.set_ylabel('Fluorescence Intensity Rank',fontsize=H*2.5)
axs.xaxis.set_ticks(np.arange(0,4501,1000))
axs.tick_params(axis='x',labelsize=H*2.5,rotation = 45)
axs.tick_params(axis='y',labelsize=H*2.5)
#ax.set_xticklabels(tick_names,rotation = 45, size=H*5, fontname='Helvetica')

axs.yaxis.set_label_position("right")
axs.yaxis.tick_right()
U1, p = mannwhitneyu( Top, Middle)
print('top middle difference',p)
U1, p = mannwhitneyu(Middle, Bottom)
print('middle down difference',p)
plt.ylim([-17, 12])
#plt.xlim([-10,100])



```
